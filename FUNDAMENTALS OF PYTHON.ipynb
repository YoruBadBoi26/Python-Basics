a tuple is another data structure
A tuple is just like a List in Python, they are both a collection of items in a variable
the major differences are as follow:
Lists are created using square brackets while Tuples are created using parentheses or also without parentheses
Lsts are mutable while Tuples are immutable i.e you cannot alter the variables in  tuple
Tuples take less memory compared to Lists
Lists have more methods(append, remove, insert, sort, etc) while Tuples have limited methods(count and index)

using the break keyword

break is used to exit the loop completely, even if the condition is still true

```python
#to count from 1 to 10, but stop when it reaches 5

count = 1

while count <= 10:
    if count ==5:
        break
    print(count)
    count += 1
```

using the continue keyword 

continue is used to skip the current iteration and move to the next

```python
#to count from 1 - 10 and skip number 5

count = 0

while count < 10:
    count += 1
    print(count)
```

```python
count = 0

while count < 10:
    count += 1
    if count == 5:
        continue
    print(count)
```

## Dealing with errors

```python
#using the try-except block
try:

    first_number = float(input("Enter the first number: "))
    second_number = float(input("Enter the second number: "))

    result = first_number/second_number
    print(result)
except ValueError:
    print("You did not enter a valid number")
except ZeroDivisionError:
    print("You cant divide by zero")
```
